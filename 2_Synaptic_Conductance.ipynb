```
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

```
sim_time = 1000 # 1 second run time
dt = 0.1 # interval size of 1x10^(-4)
time = np.arange(0, sim_time, dt)

tau = 50 # time constant
a = 1 # spike amplitude constant
R = 5/1000 # expected rate of spikes
Rdt = R*dt # expected probability of a spike in an interval

V1 = np.zeros(int(sim_time/dt)) # container for voltages
v_values = np.zeros(int(sim_time/dt))

for i in range(0, int(sim_time/dt) - 1):
    V1[i+1] = V1[i] + dt*(-V1[i]/tau + a*int(bool(Rdt > np.random.rand())))
```

```
g_l = 0.1 # leaky conductance
e_l = -60 # leaky reversal
e_s = 0 # reversal potential
c =  10 # capacitance
v_values[0] = 0
for i in range(0, int(sim_time/dt) - 1):
    v_values[i+1] = v_values[i] + (dt/c)*(g_l*(e_l-v_values[i])+100*V1[i]*(e_s - v_values[i]))
```

```
figure = plt.figure(figsize = (5, 4))
plt.plot(time, V1)
figure = plt.figure(figsize = (5, 4))
plt.plot(time, v_values)
```
